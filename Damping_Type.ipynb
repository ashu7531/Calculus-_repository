#Second Order ODE Homogenous Equation Solutions:

A second-order ODE has the form:

$$Ax'' + Bx' + Cx = 0$$
The solution of the given differential equition is given by
$$\lambda=\frac{-B\pm\sqrt{B^{2}-4AC}}{2A}$$
 
#Kinds of solution:

1. Under-Damped Case:

 When $B^{2}-4AC<0,\,\lambda_{1}\,\,and\lambda_{2}$ are complex conjugate in pairs., 

2. Critically Damped Case:

 When  $B^2 - 4AC = 0$, $\lambda_{1}\,\,and\lambda_{2}$ are real and equal


3. Over-Damped Case:

 When $B^{2}-4AC>0,\,\lambda_{1}\,\,and\lambda_{2}$ are real and distinct


Finding the roots of Given Differential Equation.

Initial conditions:-

$x(0)	=x_{0}, 
x^{'}(0)	=v_{0}$

$$x=D_{1}x_{1}+D_{2}x_{2}$$
$$x(t)=D_{1}x_{1}(t)+D_{2}x_{2}(t)$$
$$x(0)={{D_{1}e^{\lambda_{1}(0)}+D_{2}e^{\lambda_{2}(0)}}}$$
$$x_{0}=x(0)=D_{1}+D_{2}$$
$$x^{'}(t)=D_{1}x_{1}^{'}(t)+D_{2}x_{2}^{'}(t)$$
$$x^{'}(t)=D_{1}\lambda_{1}e^{\lambda_{1}t}+D_{2}\lambda_{2}e^{\lambda_{2}t}$$
$$v_{0}(0)=x^{'}(0)=D_{1}\lambda_{1}+D_{2}\lambda_{2}$$


$$\lambda=\frac{-B\pm\sqrt{B^{2}-4AC}}{2A}{{=-\frac{B}{2A}\pm\sqrt{\frac{B^{2}}{4A^{2}}-\frac{C}{A}}=-\alpha\pm\beta}}{ {\beta=\sqrt{\frac{B^{2}}{4A^{2}}-\frac{C}{A}}}}where\,D1=\frac{v_{o}-x_{o}\lambda_{2}}{\lambda_{1}-\lambda_{2}}\,\,\&\,\,D2=\frac{v_{o}-\lambda_{1}x_{o}}{\lambda_{2}-\lambda_{1}}$$


Case(i):-For Over Damping

$$x=D_{1}e^{(-\alpha+\beta)t}+D_{2}e^{(-\alpha-\beta)t}$$

Case(ii):-For Critical Damping

$$x=(x_{0}+(v_{0}-\lambda x_{0})t)e^{\frac{-B}{2A}t}$$

Case(iii):-For Under Damping

$$x=D_{1}e^{(-\alpha+i\beta)t}+D_{2}e^{(-\alpha-i\beta)t}$$

$$\Rightarrow x=e^{-\alpha t}(D_{1}e^{(i\beta)t}+D_{2}e^{(-i\beta)t})$$

```python
import matplotlib.pyplot as plt
import math
```

```python
a = float(input('Enter cofficient of 2nd derivative: '))
b = float(input('Enter cofficient of 1st derivative: '))
c = float(input('Enter cofficient of x: '))
x0 = float(input('Enter value of sol at t=0: '))
v0 = float(input('Enter value at t=0 of sol derivative: '))
 
```

```python
lambda1 = (-b - (b**2 - 4*a*c)**0.5) / (2*a)
lambda2 = (-b + (b**2 - 4*a*c)**0.5) / (2*a)
alpha = -b/(2*a)
Beta = math.sqrt(abs(b**2/(4*a**2) - (c/a)))
```

```python
# Damping Type
def Damp_type(a,b,c):
  """
  Returns : type of damping
  parameter a: 
  precondition: a must be numeric.

  parameter b:
  precondition: b must be numeric.

  parameter c:
  precondition: c must be numeric.
  """
  if b**2 - 4*a*c > 0:
    return 'Over Damped'
  elif b**2 - 4*a*c == 0:
    return 'Critically Damped'
  else:
    return 'Under Damped'
```

```python
def Graph_fun(a, b, c, x0, v0):
  Obtained_Val = Damp_type(a, b, c)
  if Obtained_Val == 'Over Damped':
    print('Over Damped')
    D_1 = (v0 - lambda2 * x0)/(lambda1 - lambda2)
    D_2 = (v0 - lambda1 * x0)/(lambda2 - lambda1)
    t_list = []
    for i in range(101):
      t_list.append(i/100)
    x_list = []
    for i in t_list:
      val = D_1 * math.exp((lambda1) * i) + D_2 * math.exp((lambda2) * i)
      x_list.append(val)
  elif Obtained_Val == 'Critically Damped':
    print('Critically Damped')
    t_list = []
    for i in range(1001):
      t_list.append(i/100)
    x_list = []
    for i in t_list:
      val = (x0 + (v0 - lambda1* x0) * i) * math.exp(lambda1 * i)
      x_list.append(val)
  else:
    print('Under Damped')
    t_list = []
    for i in range(1001):
      t_list.append(i/100)
    x_list = []
    for i in t_list:
      val = math.exp(alpha * i) * ((x0 * math.cos(Beta * i)) + (v0 + alpha * x0 / Beta) * math.sin(Beta * i))
      x_list.append(val)
  plt.plot(t_list, x_list)





  


```

```python
Graph_fun(a, b, c, x0, v0)
```
